# 모델에 적용할 신호등 학습 데이터 만들기

```python
# json 파일 읽어서 데이터 프레임에 신호등 좌표, 색 on/off, 파일명
```

```python
import json
import os
import pandas as pd
```

```python
filepath = "data/train/d_train_1920_1080_daylight_1"
classes = os.listdir(filepath)

dir_list = ['red', 'yellow', 'green']
dir_path = './data'

filename = []  # json
coor = []
dfs = []
cnt = 0
for i, cls in enumerate(classes):
    
    data =  json.load(open(filepath + "/" + cls))
    df = pd.DataFrame(data["annotation"])
   
    if not df.empty:
        # 신호등만 
        filter1 = (df['class'] == 'traffic_light')
        df = df[filter1]

        if not df.empty:
            # 자동차 신호등만 
            filter2 = (df['type'] == 'car')
            df = df[filter2]
            
            if not df.empty:
                # 3구 신호등만
                filter3 = (df['light_count'] == '3')
                df = df[filter3]
          
                if not df.empty:
                    # 가로,세로 신호등중에 가로 신호등만
                    filter4 = (df['direction'] == 'horizontal')
                    df = df[filter4]
                   
                    for i in range(0,len(df)):
                        # box, 파일이름, attribute-light
                        filename.append(cls) # 파일명
                        coor.append(df['box'].values[i]) # 좌표
                        dfs.append(df['attribute'].values[i][0]) # 색

         
```

```python
dfs
```

```python
dfs = pd.DataFrame(dfs)  # 딕셔너리 -> 데이터프레임 형태로 변환

file_data = pd.DataFrame({'파일명':filename,'좌표':coor})
data = pd.concat([file_data,dfs],axis = 1)
```

```python
data
```

```python
# 필요없는 데이터 삭제
# left_arrow 는 좌회전 화살표 표시여서 나중에 쓸듯
data = data.drop(columns =['x_light','left_arrow','others_arrow'])
```
